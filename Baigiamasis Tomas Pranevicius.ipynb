## Grafikai

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 6)
```

```python
df.groupby(['Year'])['Self-harm'].sum().plot(title='Self-harm')
```

```python
sns.barplot(x='Continent', y='Self-harm', data=df)
```

Matome kad Azija yra pirmaujantis žemynas su savižudybiu skaiciumi. Todėl galime paneigti, kad Šiaurės amerikoje daugiausiai savižudybiu

```python
sample = df.groupby(['Year'])['Self-harm'].sum().to_frame('y').reset_index()
```

```python
sample
```

```python
from sklearn.linear_model import LinearRegression
X = sample[['Year']]
y = sample['y']
model = LinearRegression()
model.fit(X, y)
```

```python
spejimas = model.predict(sample[['Year']])
```

```python
plt.scatter(X, y, color='black')
plt.plot(X.iloc[:, 0], spejimas, color='r')
tolimesni_metai = np.arange(2019, 2030).reshape(-1, 1)
plt.plot(tolimesni_metai, model.predict(tolimesni_metai), color='green')
```

Savižudybių mastas mažėja ir ateityje turėtu toliau mažėti.
